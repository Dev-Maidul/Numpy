```python
import numpy as np
arr=np.random.randint(1,100,size=(10,5)) #here row means sample and col means features
print(arr.shape)
```

```python
b=arr.reshape(5,10)
# print(b.ndim)
# print(b.shape)
print(b)
```

```python
flatten=b.flatten()
print(flatten)
column_wise_flattening=np.ravel(b,order='C')
print(column_wise_flattening)
row_wise=np.ravel(b,order='F')
print(row_wise)
```

# Concatenation

```python
a=np.random.randint(1,10,size=(2,3))
b=np.random.randint(20,30,size=(2,3))
print(a)
print(b)
```

```python
con_row=np.concatenate((a,b),axis=0) # 0 means row wise
# mane akta matrix er nise r akta matrix bosbe 
print(con_row)
```

```python
con_col=np.concatenate((a,b),axis=1) # column wise concat
# mane akta matrix er shate r akta pasha pashi boshbe 
print(a)
print(b)
print(con_col)

# row wise concetanate korar shomoi column same thakbe
# col wise concat korar shomoi row ta same thakte hobe
```
