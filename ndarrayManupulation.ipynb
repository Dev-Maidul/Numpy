# Concatenation

```python
a=np.random.randint(1,10,size=(2,3))
b=np.random.randint(20,30,size=(2,3))
print(a)
print(b)
```

```python
con_row=np.concatenate((a,b),axis=0) # 0 means row wise
# mane akta matrix er nise r akta matrix bosbe 
print(con_row)
```

```python
con_col=np.concatenate((a,b),axis=1) # column wise concat
# mane akta matrix er shate r akta pasha pashi boshbe 
print(a)
print(b)
print(con_col)

# row wise concetanate korar shomoi column same thakbe
# col wise concat korar shomoi row ta same thakte hobe
```

## arithmetic operator and mathematical functions

```python
x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4  ,5 ])
add=np.add(x,y)
print(add)
sub=x-y
print(sub)
mul=x*y
print(mul)
div=x/y
print(div)
remainder=x%y
print(remainder)
```

```python
# mathematical function
# trigonometry
# sin , cos , tan takes values in radian
sin_val=np.sin(x)
print(sin_val)
cos_val=np.cos(x)
print(cos_val)

degree_conversion=np.rad2deg(x)
print(degree_conversion)

radin_conversion=np.deg2rad(x)
print(radin_conversion)
```

```python
#log

x = np.array([ 10 , 8 , 16 , 100 ])
base_10_log_val = np.log10(x)
print(base_10_log_val)


two_diye_bhag = np.log2(x)
print(two_diye_bhag)

sqrt= np.sqrt(x)
print(sqrt)
```

```python
# different
s = np.sum(x)
print(s)

#prefix sum
cumul =np.cumulative_sum(x)
print(cumul)
```

```python
# comparison - >,< , >= ,<= , == , !=

x = np.array([ 10 , 8 , 16 , 100 ])
y = np.array([ 2 , 3 ,16  ,5 ])

greater_than = x>y

print(greater_than)

equal = x==y

print(equal)

# all () -> shob true kina
# any() -> jekono ekta true ase kina

print(np.all(greater_than))

print(np.any(equal))
```

# Sorting

```python
#inplace
x = np.array([ 10 , 8 , 16 , 100 ])
z=x.copy()
print(z)
z.sort()
print(z)
```

```python
# Copy soring
print(x)
sort_arr=np.sort(x)
print(sort_arr)
print(x)
```

```python
# 2d array sorting
mat = np.array([[10,3,5],[8,4,9]])

# horizon axis = 1

hor_sort = np.sort(mat,axis=1)
print(hor_sort)

vert_sort = np.sort(mat,axis=0)
print(vert_sort)
```

# Searching

```python
print(x)
# np.where( condition , x , y ) if true x , otherwise replace with y
index=np.where(x==8) # return index
print(index)

arr=np.where(x>8,x,0) # return array
print(arr)
```

```python
print(mat)
index=np.where(mat>8) # return index
print(index)
arr=np.where(mat>8,mat,0)
print(arr)
```

```python
x = np.array([ 10,100 , 8 , 16 , 100 ])
print(x)
max_indx=np.argmax(x)
print(max_indx)
min_idx=np.argmin(x)
print(min_idx)
```

## counting in ndarray

```python
value_greater_60=np.count_nonzero(a>60)
print(value_greater_60)
```
