Ndarray

```python
import numpy as np

#1d Array
arr1=np.array([1,2,3])

#2d Array
arr2=np.array([[1,2,3,4],[1,2,3,4]])

#3d Array
arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2


])
```

Array attributes

```python
# Check dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)
#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

```python
#set

st = { 1, 2, 3 }

arr = np.array(list(st))
# print(arr.dtype)
arr = arr.astype(np.int16)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
dc = { 'a' : 10 , 'b' : 20 , 'c' :30 }

keys = dc.keys()
values = dc.values()
items = dc.items()
print(items)
```

```python
# keys to ndarray
arr=np.array(list(keys))
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# values to ndarray
arr=np.array(list(values))
print(arr)
print(arr.size)
print(arr.shape)
print(arr.ndim)
print(arr.dtype)
```

```python
# items to ndarray
arr=np.array(list(items))
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## Creating ndarray from scratch

```python
# np.zeros
arr=np.zeros((2,3),dtype=np.int32)
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.ones
arr=np.ones((3,4,3),dtype=np.int32)
print(arr)
print(arr.shape)
print(type(arr))
print(arr.ndim)
print(arr.dtype)
print(arr.size)
```

```python
#np.empty= declare in memory but not initialize value.Later we can fill value
arr=np.empty((4,3),dtype=np.int32)
arr.fill(5)
print(arr)
print(arr.shape)
print(type(arr))
print(arr.ndim)
print(arr.dtype)
print(arr.size)
```

```python
# np.full
arr=np.full((4,3),np.inf)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

# array creation with random values

```python
# np.random.rand(shape) range here greater 0 less 1
# Every time generate a random number
arr=np.random.rand(2,3)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.random.randint(start,end, shape)
# Here everytime will generate a int number, we have to mention the starting, ending and shape of array
arr=np.random.randint(1,5,(2,3))
print(arr)
```

```python
# np.random.uniform(start,end, shape)

arr = np.random.uniform(50,100, (3,3,4))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## array creation with range functions


```python
# np.arange(first , end , step ) -> similar to range function
arr=np.arange(1,10,1)
mat=arr.reshape(3,3)
print(mat)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start , end(included) , koita)

arr = np.linspace(0, 4 ,6)


print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.logspace(start , end(included) , koita )

arr = np.logspace(0 , 4 ,6, base =2)

print(arr)
```
