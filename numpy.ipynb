Ndarray

```python
import numpy as np

#1d Array
arr1=np.array([1,2,3])

#2d Array
arr2=np.array([[1,2,3,4],[1,2,3,4]])

#3d Array
arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2


])
```

Array attributes

```python
# Check dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)
#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

# ndarry data type

```python
arr = np.array([1,2,3],dtype=np.int16)

print(arr.dtype)

arr = np.array([1.5,2.4,3.2])
print(arr.dtype)
```
