Ndarray

```python
import numpy as np

#1d Array
arr1=np.array([1,2,3])

#2d Array
arr2=np.array([[1,2,3,4],[1,2,3,4]])

#3d Array
arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2


])
```

Array attributes

```python
# Check dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)
#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

# ndarry data type

```python
arr = np.array([1,2,3],dtype=np.int16)

print(arr.dtype)

arr = np.array([1.5,2.4,3.2])
print(arr.dtype)
```

## ndarray creation from existing data ( BONUS )

```python
# list
lst=[10,23,33,21,43,22,42]
arr=np.array(lst)
# print(arr)
# print(type(arr))
# print(arr.dtype)
mixed_lst=[10,True,"Hello"]
arr=np.array(mixed_lst)
# print(arr)
# print(type(arr))
# print(arr.dtype)

# Matrix
matrix=[[1,2,3],[4,5,6]]
arr=np.array(matrix)
print(arr)
print(arr.dtype)
print(type(arr))
print(arr.shape)
print(arr.ndim)
```

```python
tpl = (10,20,30)

arr = np.array(tpl,dtype=np.int32)
print(arr.dtype)
arr = arr.astype(np.int16)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#set

st = { 1, 2, 3 }

arr = np.array(list(st))
# print(arr.dtype)
arr = arr.astype(np.int16)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
dc = { 'a' : 10 , 'b' : 20 , 'c' :30 }

keys = dc.keys()
values = dc.values()
items = dc.items()
print(items)
```

```python
# keys to ndarray
arr=np.array(list(keys))
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# values to ndarray
arr=np.array(list(values))
print(arr)
print(arr.size)
print(arr.shape)
print(arr.ndim)
print(arr.dtype)
```

```python
# items to ndarray
arr=np.array(list(items))
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## Creating ndarray from scratch

```python
# np.zeros
arr=np.zeros((2,3),dtype=np.int32)
print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.ones
arr=np.ones((3,4,3),dtype=np.int32)
print(arr)
print(arr.shape)
print(type(arr))
print(arr.ndim)
print(arr.dtype)
print(arr.size)
```
